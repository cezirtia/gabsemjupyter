```python
# Vamos carregar as bibliotecas...

import pandas as pd   # Damos o apelido de 'pd' para não ter de repetir 'pandas' 
import numpy as np    # Damos o apelido de 'np' para não ter de repetir 'numpy'
```

Depois de carregar as bibliotecas, pedimos ao Pandas para ler o arquivo de dados. Nesta análise, escolhi o arquivo `despesas.csv`.

Trata-se do __arquivo de gastos feitos por meio da verba de gabinete dos deputados estaduais__. Eu obtive este `.csv` a partir de um script de raspagem que criei -- ele não está disponível neste formato na internet.

Você pode fazer o download aqui: https://github.com/rodolfo-viana/gabsemjupyter/blob/master/data/despesas.csv.

Para subir os dados do arquivo na memória, Pandas tem um comando bem simples:

    pd.read_csv()

A documentação completa desta função está [aqui](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_csv.html), mas você não precisa saber tudo de cabeça. O importante agora é entender o funcionamento -- e como corrigir possíveis erros.

O `pd.read_csv()` lê um arquivo e converte em `DataFrame`. Por exemplo:

    pd.read_csv('despesas.csv')

Mas se eu digitar isso, terei um erro:

```python
dataset = pd.read_csv('despesas.csv')
```

O primeiro erro que encontramos foi este:

    FileNotFoundError: File b'despesas.csv' does not exist

Motivo: o arquivo não está no mesmo diretório deste notebook. Ele está na pasta `data`.

O correto, portanto, é indicar onde está o arquivo:

    pd.read_csv('data/despesas.csv')

Mas, se eu digitar isso, terei erro novamente -- outro erro.

```python
dataset = pd.read_csv('data/despesas.csv')
```

Repare que o tipo de erro mudou: em vez de não encontrar o arquivo, o notebook não conseguiu lê-lo devido a um erro de codificação.

    UnicodeDecodeError: 'utf-8' codec can't decode byte 0xc1 in position 4: invalid start byte

Como padrão, Pandas tenta a leitura usando a codificação `utf-8`. Mas nosso arquivo está com outra codificação. Repare no que acontece quando eu o abro num editor de planilhas:

![Encoding](https://raw.githubusercontent.com/rodolfo-viana/gabsemjupyter/master/img/6613531705999360.png "Encoding")

Vê as interrogações nos losangos no meio das palavras? Isso indica que nosso arquivo não está em UTF-8.

Para o Pandas poder ler, é preciso indicar a codificação correta. Isso é feita com `encoding`:

    pd.read_csv('data/despesas.csv', encoding = 'latin-1')

Latin-1 é o apelido do ISO-8859-1. E o que isso quer dizer? Em uma frase captada no [Stack Overflow](https://stackoverflow.com/questions/2708958/differences-between-utf8-and-latin1):

    "UTF-8 is prepared for world domination, Latin-1 isn't."

Para explicar de maneira resumida, UTF-8 contém uma variedade maior de caracteres, enquanto Latin-1 é adequada apenas para línguas derivadas do latim -- e no latim em si não havia cedilha, por exemplo. Daí vem o erro. (E por isso o Pandas vem "de fábrica" preparado para ler UTF-8.)

Vamos testar o comando com o `encoding` definido:

```python
dataset = pd.read_csv('data/despesas.csv', encoding = 'latin-1')
```

Agora funcionou -- mas com um aviso:

    DtypeWarning: Columns (5) have mixed types. Specify dtype option on import or set low_memory=False.

É um erro de tipo. Nosso dataset tem `string`, `integer` e `float`, e o trabalho de armazenar na memória essa variedade toda precisa ser maior. Por isso, há dois caminhos:

    1. Definir os tipos de cada coluna, ou
    2. Pedir para usar mais memória (ou seja, 'low_memory = False').

Vamos da segunda opção, e depois faremos a primeira:

```python
dataset = pd.read_csv('data/despesas.csv', encoding = 'latin-1', low_memory = False)
```

Pronto! Nenhum erro!

Vamos testar:

```python
dataset
```

Caso você preferisse indicar o tipo das colunas, poderia usar Numpy, e o código ficaria assim:

```python
dataset = pd.read_csv('data/despesas.csv',           # Fui para a linha de baixo para dar leitura
                      encoding = 'latin-1',            
                      dtype = {'Deputado': np.str,
                               'Matricula': np.int,
                               'Ano': np.int,
                               'Mes': np.int,
                               'Tipo': np.str,
                               'CNPJ': np.str,       # Como não vou fazer soma, trato CNPJ como 'str'
                               'Fornecedor': np.str,
                               'Valor': np.float})
```

Mesmo sem `memory_low = False`, nenhum erro.

Agora que já temos o dataset na memória -- com o nome `dataset` --, vamos ver os dados.

```python
dataset
```

Repare que, além das oito colunas do nosso dataset, o Pandas criou um índice na esquerda -- que começa no 0 (zero). Isso é bastante útil em muitas ocasiões: sabemos, por exemplo, que estamos trabalhando com 446.408 registros (neste caso, notas fiscais).

E por haver tantos registros, o Pandas esconde alguns -- ou teríamos uma lista imensa cuja leitura seria praticamente inútil. Repare que são apresentados os 30 primeiros registros, reticências, e os 30 últimos.
