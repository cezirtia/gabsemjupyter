Agora que vimos como documentar um notebook, salvar na memória o dataset e dar uma primeira olhada nos dados, é hora de colocar a mão na massa!

- responder a uma pergunta pré-definida;

Mas, por experiência, navegar no meio de números sem propósito também pode render boas histórias. A matéria dos gastos dos vereados de São Paulo com correios nasceu assim, por exemplo.

A verdade é que uma origem não precisa necessariamente eliminar a outra: você pode tentar responder a uma pergunta pré-estabelecida e, ao mesmo tempo, manter a mente aberta para outras coisas que encontrar no caminho.

Para exercitar, bamos subir o primeiro dataset, de despesas dos deputados:

```python
# Importação das ferramentas
import pandas as pd
import numpy as np

# Gravação de dataset de despesas na memória
despesas = pd.read_csv('data/despesas.csv', 
                       encoding = 'latin-1',
                       dtype = {'Deputado': np.str,
                                'Matricula': np.int,
                                'Ano': np.int,
                                'Mes': np.int,
                                'Tipo': np.str,
                                'CNPJ': np.str, 
                                'Fornecedor': np.str,
                                'Valor': np.float})

# Resultado
despesas.shape
```

Com o dataset gravado na memória, cabe ressaltar uma das boas práticas da análise de dados: **não altere o dataset original!**

Por uma questão de lisura, é recomendado deixar o dataset original intacto, sem alteração dos registros.

O que você pode fazer é "batizá-lo" de outras formas. Por exemplo, se eu quero apenas os dados referentes a 2017:

```python
# Recorte de 2017
despesas_2017 = despesas[despesas['Ano'] == 2017]

# Resultado
despesas_2017.shape
```

Ou se eu quero os gastos com a categoria "N - MORADIA":

```python
# Recorte de 2017
despesas_moradia = despesas[despesas['Tipo'] == 'N - MORADIA']

# Resultado
despesas_moradia.shape
```

Repare que, tanto no recorte de ano quanto no de tipo, fiz referência ao dataset original -- afinal, ele é a fonte --, mas não o alterei e, por isso, posso retornar a ele sempre que necessário:

```python
# Resultado do dataset original
despesas.shape
```

Vamos ver de maneira mais aprofundada o que temos no dataset original:

```python
# Estatísticas
despesas.describe()
```

Parece confuso... Mas não é. Juro.

`DataFrame.describe()` mostra algumas estatísticas das colunas que tiverem valores `float` ou `integer` -- ou seja, tudo que for número.

Esta função mostra:

`count`: a contagem de registros na coluna -- no dataset em questão, 446.408 registros ou linhas.<br>
`mean`: a média dos valores de cada coluna -- o que não faz muito sentido para "Matricula", "Ano" ou "Mes", mas faz para "Valor", pois sabemos que o valor médio de todos os registros é 680.26 reais.<br>
`std`: o desvio-padrão, ou seja, a dispersão dos números. O desvio-padrão pequeno indica que os valores estão mais concentrados em torno da média. Exemplo:

`15 + 25 + 20 = 60 => os números estão mais concentrados em torno do meio; logo, desvio_padrão pequeno.`<br>
`05 + 51 + 04 = 60 => os números estão mais dispersos em relação ao meio; logo, desvio-padrão grande.`

`min`: o menor número de cada coluna. No nosso dataset, o menor ano é 2002, e o menor valor é 0 real.<br>
`25%`, `50%` e `75%`: respectivamente o quartil inferior, a mediana e o quartil inferior. Em termos gerais, mostra os valores (ordenados) contidos nos primeiros 25% registros da amostra, na metade da amostra e nos 75% registros da amostra. No nosso dataset, 25% dos valores estão até 50 reais, a mediana é 127 reais, e 75% estão até 495,04 reais.<br>
`max`: o número máximo em cada coluna. No nosso caso, o ano máximo é 2018, e o valor máximo é 13.262,00.

Sabendo isso, a função `DataFrame.describe()` revela, entre outras coisa, que, no nosso dataset:

- há 446.408 registros
- de 2002 a 2018
- com valor médio de 680.26 reais
- sendo o valor mínimo de 0 real
- e o valor máximo de 13.262,00

Mas se ainda parecer confuso -- afinal, não existe um desvio-padrão em mês, ou um valor mínimo em matrícula --, basta selecionar uma coluna única:

```python
# Estatísticas de valores
despesas['Valor'].describe()
```

E quanto aos recortes que fizemos? Vamos ver:

```python
# Estatísticas de valores de 2017
despesas_2017['Valor'].describe()
```

```python
# Estatísticas de valores de moradia
despesas_moradia['Valor'].describe()
```

Ok. Temos um recorte de 2017 e outro de moradia. Mas e se quisermos apenas um recorte contemplando os dois critérios? É bastante simples...

Se a lógica de recorte é:

    dataset_secundario = dataset[dataset['coluna'] == 'criterio']

Então, com dois critérios, é:

    dataset_secundario = dataset[(dataset['coluna'] == 'criterio 1') & (dataset['coluna'] == 'criterio 2')]

Vamos testar:

```python
# Recorte com dois critérios
despesas_duplo_recorte = despesas[(despesas['Ano'] == 2017) & (despesas['Tipo'] == 'N - MORADIA')]

# Resultado
despesas_duplo_recorte.head(20)
```

Funcionou!

Mas a escolha de um recorte vai além disso: em vez de você usar `==` para buscar por valores que atendam exatamente aos critérios estabelecidos, pode usar:

`<`: menor que (em que se exclui o valor)<br>
`<=`: igual ou menor que (em que se inclui o valor)<br>
`>`: maior que (em que se exclui o valor)<br>
`>=`: igual ou maior que (em que se inclui o valor)<br>
`!=`: todos os valores, exceto

Vamos testar. Vamos ver quantos gastos de 2017 com valor superior a 5.000 reais existem:

```python
# Recorte com dois critérios
despesas_2017_acima_5k = despesas[(despesas['Ano'] == 2017) & (despesas['Valor'] > 5000)]

#Resultado
despesas_2017_acima_5k.shape
```

Agora vamos ver em que categorias  estão esses 1.263 registros. Para isso, vamos usar uma função que conta a quantidade de registros:

    DataFrame.value_counts()

```python
# Contagem de registros por tipo
despesas_2017_acima_5k['Tipo'].value_counts()
```

Notamos que a categoria de serviços gráficos contém boa parte das notas acima de 5.000 reais.

Mas quantas notas temos de cada categoria em todo o ano de 2017, descartando o recorte de valor acima de 5.000 reais? 

Bem, já segmentamos a base de ano -- `despesas_2017` -- linhas acima. Vamos usá-la e aplicar a mesma função:

```python
# Contagem de registros por tipo
despesas_2017['Tipo'].value_counts()
```

Descobrimos agora que boa parte dos registros são gastos com combustíveis. Quanto será que eles somam?

Para encontrar a resposta, vamos segmentar o dataset `despesas_2017` e usar uma outra função:

    DataFrame.sum()

```python
# Recorte do dataset por tipo
despesas_2017_tanque = despesas_2017[despesas_2017['Tipo'] == 'A - COMBUSTÍVEIS E LUBRIFICANTES']

# Soma dos valores
despesas_2017_tanque['Valor'].sum()
```

Chegamos a algo interessante: em 2017, os deputados estaduais gastaram 2.067.121,86 com combustível.

Mas esse é o valor total. Qual é a média de gasto com combustível por nota fiscal? Para isso, podemos usar a função `DataFrame.describe()` vista linhas acima, ou ser mais direto, usando uma função que retorna apenas a média:

    DataFrame.mean()

Vamos fazer dos dois jeitos:

```python
# Estatísticas dos valores
despesas_2017_tanque['Valor'].describe()
```

```python
# Média dos valores
despesas_2017_tanque['Valor'].mean()
```

Vemos que a média por nota fiscal é 311,64.

Ah, há alguns números além dos dois decimais a que estamos acostumados. Isso porque o Pandas não sabe que estamos trabalhando com valor monetário -- ele é usado em diversas áreas que contemplam mais de duas casas decimais.

Logo você se habitua a automaticamente arrendondar para mais ou para menos, a depender do valor. Mas, se fizer questão de usar apenas dois dígitos, pode recorrer à função:

    DataFrame.round(decimals = 2)

```python
# Arredondamento da média dos valores em dois dígitos
despesas_2017_tanque['Valor'].mean().round(decimals = 2)
```
